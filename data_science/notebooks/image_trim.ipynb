```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
stereo = cv.StereoBM_create(numDisparities=16, blockSize=15)
imgL = cv.imread('../data/tsukuba_l.png',0)
imgR = cv.imread('../data/tsukuba_r.png',0)
```

```python
disparity = stereo.compute(imgL,imgR)
```

```python
plt.imshow(disparity,'gray')
plt.show()
```

```python
h = 16
y = 125
imgL_trim = imgL[y:y+h].copy()
imgR_trim = imgR[y:y+h].copy()
```

```python
disparity = stereo.compute(imgL_trim, imgR_trim)
```

```python
plt.imshow(imgL_trim)
plt.show()
plt.imshow(disparity, 'gray')
plt.show()
```

```python
for arr in range(len(disparity)):
    for i in disparity[arr]:
        if i != -16:
            print(arr)
            break
```
