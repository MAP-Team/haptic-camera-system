```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
from imutils.video import WebcamVideoStream
import imutils
```

```python
vs1 = WebcamVideoStream(src=1).start()
vs2 = WebcamVideoStream(src=2).start()     
```

```python
imgR = cv2.cvtColor(imutils.resize(vs1.read(), width=400), cv2.COLOR_BGR2GRAY)
imgL = cv2.cvtColor(imutils.resize(vs2.read(), width=400), cv2.COLOR_BGR2GRAY)
```

```python
stereo = cv2.StereoBM_create(numDisparities=16, blockSize=15)
```

```python
plt.imshow(imgL)
```

```python
plt.imshow(imgR)
```

```python
disparity = stereo.compute(imgL,imgR)
```

```python
plt.imshow(disparity,'gray')
plt.show()
```
